**Metric Evaluation**

Dalam evaluasi model machine learning untuk prediksi promosi karyawan, terdapat dua jenis kesalahan yang perlu diperhatikan:

- **Type 1 Error (False Positive)**: Karyawan yang diprediksi akan dipromosikan, tetapi sebenarnya tidak layak dipromosikan. 

- **Type 2 Error (False Negative)**: Karyawan yang sebenarnya layak promosi tetapi tidak terdeteksi oleh model. 

Untuk mengevaluasi akurasi model dalam prediksi kelayakan promosi, beberapa metrik digunakan, dengan F-Beta Score dan Precision sebagai prioritas untuk dapat mendeteksi karyawan yang layak promosi secara optimal:

1. **F1-score**: Rata-rata harmonis antara precision dan recall. Metrik ini diprioritaskan karena menjaga keseimbangan dalam meminimalkan False Positives dan False Negatives, sehingga karyawan yang layak promosi teridentifikasi tanpa membebani tim HR dengan kandidat yang tidak relevan.

2. **Precision**: Mengukur ketepatan model dalam memprediksi promosi, yaitu berapa banyak dari seluruh prediksi promosi yang benar-benar layak dipromosikan. Precision membantu mengurangi False Positives.

3. **Recall**: Mengukur kemampuan model dalam mendeteksi semua karyawan yang benar-benar layak promosi, sehingga mengurangi False Negatives.

4. **ROC-AUC**: Mengukur kemampuan model dalam memisahkan kelas promosi dan non-promosi, memberikan gambaran menyeluruh tentang performa model dalam membedakan dua kelas tersebut.

5. **F-Beta Score**: Mengukur keseimbangan antara Precision dan Recall dengan menyesuaikan tingkat kepentingan masing-masing. Beta adalah parameter yang menentukan prioritas:

    Beta < 1: Fokus lebih pada Precision.
    Beta = 1: Memberikan bobot yang sama antara Precision dan Recall (setara dengan F1-Score).
    Beta > 1: Memberikan prioritas lebih pada Recall.

Konsekuensi Type 1 dan Type 2 Error
1. **Dampak False Negatives**
    
    Karyawan yang sebenarnya layak promosi namun tidak terdeteksi oleh model merasa tidak dihargai, yang dapat berdampak pada tingkat kepuasan mereka. Jika mereka keluar dari perusahaan, biaya untuk merekrut dan melatih karyawan eksternal dapat mencapai 18-20% lebih tinggi dibandingkan biaya untuk karyawan internal. Menghindari terlalu banyak False Negatives penting untuk mempertahankan karyawan berbakat. (Referensi biaya perekrutan karyawan: https://knowledge.wharton.upenn.edu/article/why-external-hires-get-paid-more-and-perform-worse-than-internal-staff/). Namun dikarenakan dampak false negative ini sulit dihitung secara langsung, maka akan lebih diprioritaskan kepada dampak false positive.

2. **Dampak False Positives**
    
    Jika terlalu banyak False Positives (misalnya, karyawan yang sebenarnya tidak layak promosi tetapi direkomendasikan oleh model), tim HR akan menghabiskan waktu dan sumber daya untuk mengevaluasi karyawan yang tidak relevan. Estimasi biaya evaluasi per karyawan adalah $240. Jika dari 100 prediksi promosi hanya 70 yang benar-benar layak, maka $7,200 (30 karyawan x $240) akan terbuang untuk evaluasi yang tidak produktif. (Referensi biaya evaluasi karyawan: https://zensai.com/articles/traditional-performance-review-stats/)

Type 1 Error (False Positive) memiliki dampak yang lebih langsung dan terukur dibandingkan dengan Type 2 Error (False Negative). Hal ini disebabkan oleh konsekuensi yang jelas dari proses evaluasi yang melibatkan kandidat yang tidak relevan, sehingga meningkatkan biaya dan waktu evaluasi. Oleh karena itu, Precision diprioritaskan untuk meminimalkan kandidat tidak relevan, memastikan proses evaluasi lebih efisien, dan mengoptimalkan sumber daya tim HR.

Namun, dampak Type 2 Error (False Negative) tetap tidak bisa diabaikan. Ketika karyawan yang layak promosi tidak terdeteksi oleh model, mereka kehilangan kesempatan untuk mengikuti evaluasi lanjutan. Hal ini dapat menyebabkan karyawan merasa tidak dihargai, yang berpotensi meningkatkan ketidakpuasan, mengurangi retensi, dan meningkatkan turnover karyawan berbakat.

Mengacu pada kebutuhan spesifik model ini yang berfungsi sebagai alat bantu penyaringan awal, dan bukan sebagai keputusan akhir untuk promosi karyawan, penggunaan F-Beta Score adalah solusi yang ideal. Dengan memberikan bobot lebih pada Precision melalui parameter Beta, perusahaan dapat menjaga keseimbangan antara pengurangan False Positives dan dampak dari False Negatives.

### **Outlier**

Pertama, outlier akan di deteksi dengan cara univariate:

```python
# Boxplot untuk outliers
plt.figure(figsize=(20,6))
sns.boxplot(data=df.drop(columns=['employee_id','department','region','education','gender','recruitment_channel','awards_won?','is_promoted']))
plt.title('Boxplot of Numerical Features')
plt.show()
```

```python
numerical = df.drop(columns=['employee_id','department','region','education','gender','recruitment_channel','awards_won?','is_promoted']).columns

def calculate_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

outlier_data = []

for column in numerical:
    outlier_result, lower_bound, upper_bound = calculate_outliers(df, column)
    total_outlier = len(outlier_result)
    outlier_percentage = round(total_outlier / len(df[column]) * 100, 2)
    outlier_data.append([column, total_outlier, outlier_percentage, lower_bound, upper_bound])

outlier_df = pd.DataFrame(outlier_data, columns=["Column", "Total Outliers", "Percentage (%)", "Lower Bound", "Upper Bound"])
outlier_df
```

Terlihat bahwa terdapat beberapa data yang termasuk outlier pada setiap fiturnya. Namun, dikarenakan jumlah nya yang banyak, outlier tidak akan dihapuskan dari pembuatan model machine learning.

Pertimbangan tidak membuang outlier adalah dikarenakan beberapa faktor yaitu:
- Outlier tidak disebabkan oleh kesalahan pencatatan atau data yang tidak valid. Terlihat pada boxplot, persebaran outlier masih berkumpul dan tidak ada yang sangat jauh dari yang lainnya. Selain itu, tidak ada angka yang tidak mungkin yaitu angka negatif untuk setiap fiturnya.
- Ada kemungkinan data outlier mencerminkan karakteristik karyawan yang valid (misalnya, karyawan yang sudah bekerja lama dan umurnya diatas rata-rata karyawan lainnya serta sudah mendekati umur pensiun), mereka sebaiknya tidak dihapus karena mereka adalah bagian dari data yang relevan.
- Selain model regresi yang sensitif terhadap outlier, model machine learning juga akan dicoba dengan model yang biasanya tidak terlalu terpengaruh terhadap outlier, seperti decision tree, random forest, atau boosting. Dalam kasus ini, outlier tidak harus dihapus.

Selain itu, untuk mengurangi pengaruh outlier, akan digunakan scalling seperti Robust Scaling (berpengaruh untuk regresi logistik) pada setiap fitur numerik yang akan membuat model lebih tahan terhadap outlier.

### **Numerical & Categorical Data Analysis**

```python
def calculate_percentage(series):
    counts = series.value_counts()
    percentages = (counts / counts.sum()) * 100
    return percentages

# Function to add labels to bars
def add_labels_1(ax):
    for container in ax.containers:
        ax.bar_label(container, fmt='%.1f%%', label_type='edge', padding=3)

def add_labels_2(ax):
    for container in ax.containers:
        ax.bar_label(container, fmt='%.2f%%', label_type='edge', padding=3)

def add_labels_0(ax):
    for container in ax.containers:
        ax.bar_label(container, fmt='%.0f%%', label_type='edge', padding=3)
```

**Barplot of no_of_trainings**

Distribusi jumlah pelatihan menunjukkan bahwa sebagian besar karyawan hanya mengikuti 1 atau 2 pelatihan, dengan distribusi yang sangat condong ke kanan.

2. Pengisian data yang hilang. Cara ini lebih disarankan dan ada beberapa metode untuk mengisi missing value tersebut. Cara yang terbaik adalah dengan mengisi data kosong dengan data yang mendekati dengan data sebenarnya. Dalam kasus ini, data kosong akan akan diisi secara domain knowledge atau secara statistik berdasarkan dengan kolom lainnya. Jika masih ada kolom yang tidak bisa diisi, maka akan diisi dengan angka mean, median, modus atau algoritma lainnya. Menghapus data adalah pilihan terakhir.

### Mengisi Data Kosong kolom categorical dengan cara modus

Mengisi data kosong dengan **modus** berarti mengganti nilai yang hilang dalam kolom data dengan nilai yang paling sering muncul (modus) di kolom tersebut. Teknik ini umum digunakan untuk data kategorikal agar distribusi data tetap konsisten. Langkahnya adalah mengidentifikasi kolom yang memiliki data kosong, menghitung nilai modus dari kolom tersebut, lalu mengganti semua data kosong dengan nilai modus. Cara ini sederhana dan efektif untuk menjaga integritas data tanpa menambahkan ketidakteraturan. Namun, metode ini kurang sesuai untuk data numerik, di mana median atau rata-rata lebih sering digunakan.

```python
df["education"].value_counts()
```

Sebelum dilanjutkan pengisian data dengan cara Modus, akan dilihat kembali apakah kolom education masih memiliki korelasi antara fitur lain yaitu salah satunya fitur umur.

```python
# Calculate the proportion of education levels within each age group
education_age_proportion = df.groupby(['age', 'education']).size().unstack(fill_value=0)
education_age_proportion = education_age_proportion.div(education_age_proportion.sum(axis=1), axis=0)

# Display the resulting table
print(education_age_proportion)
```

```python
# Plotting
education_age_proportion.plot(kind='line', marker='o', figsize=(10, 6))

plt.title('Proportion of Education Levels by Age Group')
plt.xlabel('Age')
plt.ylabel('Proportion')
age_min = df['age'].min()
age_max = df['age'].max()
plt.xticks(range(age_min, age_max + 1, 1))
plt.legend(title='Education Level')
plt.grid(True)
plt.show()
```

### Mengisi data kosong kolom numerikal dengan cara median
Mengisi data kosong dengan **median** berarti mengganti nilai yang hilang dalam kolom data numerik dengan nilai tengah (median) dari kolom tersebut. Langkah ini berguna agar data tetap mewakili distribusi asli, terutama jika ada outlier yang bisa memengaruhi rata-rata. Cara mengisinya adalah dengan mengidentifikasi kolom yang memiliki data kosong, menghitung median dari kolom tersebut, lalu mengganti setiap data kosong dengan nilai median. Metode ini lebih cocok untuk data numerik dibandingkan modus, karena median lebih stabil dan tidak dipengaruhi oleh nilai ekstrem.

Dari matriks korelasi antara fitur, tidak terdapat hubungan kuat antara fitur ini dengan fitur lainnya, makan akan dilanjutkan dengan pengisian data kosong dengan nilai median nya.

Logistic Regression juga memberikan interpretasi yang jelas tentang pengaruh fitur melalui koefisien model.

2. **KNeighbors Classifier (KNN)**: KNN adalah metode klasifikasi non-parametrik yang bekerja berdasarkan kesamaan antara data. Pada dasarnya, model ini mengklasifikasikan sampel baru berdasarkan mayoritas label dari tetangga terdekatnya dalam ruang fitur. Untuk pemilihan model, akan dilakukan cross validation pada model dengan deafult parameter terlebih dahulu.  Untuk metode cross-validation akan dipilih dengan metode **Stratified K-Fold** dikarenakan metode ini memastikan bahwa setiap fold dari data memiliki proporsi yang sama untuk kelas yang dipromosi dan tidak dipromosi. 

Dalam setiap fold, Stratified K-Fold menjaga distribusi 9.5% dipromosi dan 90.5% dipromosi, sehingga model dapat dilatih dan divalidasi pada set data yang lebih representatif dari keseluruhan populasi.

```python
from imblearn.pipeline import Pipeline
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.metrics import make_scorer, fbeta_score

# Define the list of models
models = [logreg, knn, dt, rf, xgb, lgbm]

# Initialize lists to store scores
score_fbeta = []
mean_fbeta = []
std_fbeta = []
score_precision = []
mean_precision = []
std_precision = []

transformer = ColumnTransformer([
    ('scaling', RobustScaler(), ['no_of_trainings', 'age', 'length_of_service', 'avg_training_score']),
    ('onehot', OneHotEncoder(drop='first'), ['education', 'gender', 'recruitment_channel']),
    ('binary',BinaryEncoder(),['department','region'])
], remainder='passthrough')

# Define F-Beta Scorer with focus on Precision
beta = 0.3  # Beta < 1 to focus on Precision
fbeta_scorer = make_scorer(fbeta_score, beta=beta, average='binary')

# Loop through each model
for i in models:
    skfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
    estimator = Pipeline([
        ('preprocess', transformer),
        ('model', i)
    ])
    model_cv_fbeta = cross_val_score(estimator, x_train, y_train, cv=skfold, scoring=fbeta_scorer)
    score_fbeta.append(model_cv_fbeta)
    mean_fbeta.append(model_cv_fbeta.mean())
    std_fbeta.append(model_cv_fbeta.std())

    model_cv_precision = cross_val_score(estimator, x_train, y_train, cv=skfold, scoring='precision')
    score_precision.append(model_cv_precision)
    mean_precision.append(model_cv_precision.mean())
    std_precision.append(model_cv_precision.std())

# Create a DataFrame with Precision score results
fbeta_scores_df = pd.DataFrame({
    'Model': ['Logistic Regression', 'KNN', 'Decision Tree', 'Random Forest', 'XGBoost', 'LightGBM'],
    'Mean F-Beta Score': mean_fbeta,
    'Std F-Beta Score': std_fbeta,
    'Mean Precision Score': mean_precision,
    'Std Precision Score': std_precision
}).set_index('Model').sort_values(by=['Mean F-Beta Score'], ascending=False)

# Display the results
fbeta_scores_df
```

Berdasarkan keseluruhan evaluasi metrik pada data training, model LightGBM memiliki nilai F-Beta Score (beta=0.3) dan Precision yang paling tinggi dibandingkan model lainnya. Dengan demikian, model ini menjadi pertimbangan utama dalam pemilihan model yang terbaik.

Maka akan dilakukan teknik Oversampling dikarenakan:
1. Menyediakan lebih banyak data untuk kelas minoritas, model dapat beradaptasi lebih baik dan menghasilkan prediksi yang lebih akurat untuk kedua kelas.
2. Untuk model yang ditujukan untuk mendeteksi karyawan layak promosi, f1 score adalah metrik yang penting. Oversampling membantu meningkatkan recall dengan memberikan lebih banyak contoh positif untuk model belajar.

Oversampling akan dilakukan dengan metode **RandomOverSampler** dimana metode ini akan secara sederhana membuat salinan data dari kelas minoritas secara acak untuk meningkatkan jumlah contoh kelas minoritas hingga seimbang dengan kelas mayoritas.

Untuk metode SMOTE tidak dilakukan pada dataset. Walaupun bisa mengurangi kemungkinan overfitting dan menambah variasi pada kelas minoritas dengan data yang lebih beragam. Namun, metode ini lebih efektif untuk dataset yang besar dan dapat menghasilkan data sintetis yang tidak realistis jika salah satu contoh kelas minoritas yang digunakan adalah outlier. Selain itu, metode ini sulit diterapkan untuk fitur kategorikal dimana dataset terdapat beberapa fitur kategorikal

Oversampling hanya akan dilakukan pada data training. Dibawah ini akan dicoba terlebih dahulu apakah betul dengan metode oversampling akan menjadikan model lebih baik.

```python
from imblearn.over_sampling import RandomOverSampler
from sklearn.metrics import classification_report, roc_auc_score, accuracy_score, f1_score, recall_score, precision_score
```

```python
data = x_train
target = y_train


skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=0)

transformer = ColumnTransformer([
    ('scaling', RobustScaler(), ['no_of_trainings', 'age', 'length_of_service', 'avg_training_score']),
    ('onehot', OneHotEncoder(drop='first'), ['education', 'gender', 'recruitment_channel']),
    ('binary', BinaryEncoder(), ['department', 'region'])
], remainder='passthrough')

# Untuk menyimpan hasil evaluasi
results_ros = []
results_no_oversampling = []

# Loop melalui setiap fold
for train_index, test_index in skf.split(data, target):
    x_train_fold, x_test_fold = data.iloc[train_index], data.iloc[test_index]
    y_train_fold, y_test_fold = target.iloc[train_index], target.iloc[test_index]

    # Model tanpa oversampling
    lgbm_no_oversampling = lgb.LGBMClassifier()
    estimator_no_oversampling = Pipeline([
        ('preprocess', transformer),
        ('model', lgbm_no_oversampling)
    ])
    estimator_no_oversampling.fit(x_train_fold, y_train_fold)

    # Evaluasi untuk data pelatihan
    y_pred_no_oversampling_train = estimator_no_oversampling.predict(x_train_fold)
    y_predict_proba_no_oversampling_train = estimator_no_oversampling.predict_proba(x_train_fold)[:, 1]

    # Simpan hasil untuk model tanpa oversampling di data pelatihan
    results_no_oversampling.append({
        'roc_auc_train': roc_auc_score(y_train_fold, y_predict_proba_no_oversampling_train),
        'precision_train': precision_score(y_train_fold, y_pred_no_oversampling_train),
        'recall_train': recall_score(y_train_fold, y_pred_no_oversampling_train),
        'f_beta_train': fbeta_score(y_train_fold, y_pred_no_oversampling_train, beta=0.3),  # F-beta with beta=0.5
    })

    # Evaluasi untuk data pengujian
    y_pred_no_oversampling_test = estimator_no_oversampling.predict(x_test_fold)
    y_predict_proba_no_oversampling_test = estimator_no_oversampling.predict_proba(x_test_fold)[:, 1]

    # Simpan hasil untuk model tanpa oversampling di data pengujian
    results_no_oversampling[-1].update({
        'roc_auc_test': roc_auc_score(y_test_fold, y_predict_proba_no_oversampling_test),
        'precision_test': precision_score(y_test_fold, y_pred_no_oversampling_test),
        'recall_test': recall_score(y_test_fold, y_pred_no_oversampling_test),
        'f_beta_test': fbeta_score(y_test_fold, y_pred_no_oversampling_test, beta=0.3),  # F-beta with beta=0.5
    })

    # Random Oversampling
    random_over_sampler = RandomOverSampler(random_state=0)
    x_train_ros, y_train_ros = random_over_sampler.fit_resample(x_train_fold, y_train_fold)

    # Model dengan Random Oversampling
    lgbm_ros = lgb.LGBMClassifier()
    estimator_ros = Pipeline([
        ('preprocess', transformer),
        ('model', lgbm_ros)
    ])
    estimator_ros.fit(x_train_ros, y_train_ros)

    # Evaluasi untuk data pelatihan
    y_pred_ros_train = estimator_ros.predict(x_train_ros)
    y_predict_proba_ros_train = estimator_ros.predict_proba(x_train_ros)[:, 1]

    # Simpan hasil untuk model dengan Random Oversampling di data pelatihan
    results_ros.append({
        'roc_auc_train': roc_auc_score(y_train_ros, y_predict_proba_ros_train),
        'precision_train': precision_score(y_train_ros, y_pred_ros_train),
        'recall_train': recall_score(y_train_ros, y_pred_ros_train),
        'f_beta_train': fbeta_score(y_train_ros, y_pred_ros_train, beta=0.3),  # F-beta with beta=0.5
    })

    # Evaluasi untuk data pengujian
    y_pred_ros_test = estimator_ros.predict(x_test_fold)
    y_predict_proba_ros_test = estimator_ros.predict_proba(x_test_fold)[:, 1]

    # Simpan hasil untuk model dengan Random Oversampling di data pengujian
    results_ros[-1].update({
        'roc_auc_test': roc_auc_score(y_test_fold, y_predict_proba_ros_test),
        'precision_test': precision_score(y_test_fold, y_pred_ros_test),
        'recall_test': recall_score(y_test_fold, y_pred_ros_test),
        'f_beta_test': fbeta_score(y_test_fold, y_pred_ros_test, beta=0.3),  # F-beta with beta=0.5
    })

# Menghitung rata-rata untuk semua metrik
avg_metrics_no_oversampling = pd.DataFrame(results_no_oversampling).mean().to_dict()
avg_metrics_ros = pd.DataFrame(results_ros).mean().to_dict()

# Membuat DataFrame untuk hasil metrik yang diminta
metrics_df = pd.DataFrame({
    'Metrics': ['roc_auc', 'precision', 'recall', 'f_beta'],
    'No Oversampling Train': [
        avg_metrics_no_oversampling['roc_auc_train'],
        avg_metrics_no_oversampling['precision_train'],
        avg_metrics_no_oversampling['recall_train'],
        avg_metrics_no_oversampling['f_beta_train']
    ],
    'No Oversampling Test': [
        avg_metrics_no_oversampling['roc_auc_test'],
        avg_metrics_no_oversampling['precision_test'],
        avg_metrics_no_oversampling['recall_test'],
        avg_metrics_no_oversampling['f_beta_test']
    ],
    'Random Oversampling Train': [
        avg_metrics_ros['roc_auc_train'],
        avg_metrics_ros['precision_train'],
        avg_metrics_ros['recall_train'],
        avg_metrics_ros['f_beta_train']
    ],
    'Random Oversampling Test': [
        avg_metrics_ros['roc_auc_test'],
        avg_metrics_ros['precision_test'],
        avg_metrics_ros['recall_test'],
        avg_metrics_ros['f_beta_test']
    ]
}).set_index('Metrics')

# Set index dan tampilkan DataFrame
metrics_df
```

**Hyperparameter tuning**

Hyperparameter tuning adalah proses mencari kombinasi nilai optimal dari hyperparameter suatu model untuk meningkatkan accuracynya pada data tertentu. Berbeda dengan parameter model (seperti koefisien regresi), hyperparameter tidak dipelajari dari data, melainkan ditentukan sebelum pelatihan. Proses tuning bertujuan untuk mengoptimalkan performa model pada metrik tertentu (misalnya, Precision Score atau ROC AUC).

Ada dua pendekatan umum untuk hyperparameter tuning:

1. Grid Search: Menguji semua kemungkinan kombinasi hyperparameter dalam ruang pencarian (parameter grid). Pendekatan ini komprehensif tetapi bisa memakan waktu jika grid besar.

2. Random Search: Memilih subset kombinasi hyperparameter secara acak. Lebih efisien untuk grid besar tetapi mungkin melewatkan kombinasi optimal.

Kita akan memakai Grid Search karena memiliki banyak manfaat yang membuatnya menjadi metode yang andal dalam melakukan tuning hyperparameter. Grid Search melakukan evaluasi menyeluruh terhadap semua kombinasi hyperparameter yang mungkin dalam ruang pencarian yang telah ditentukan. Hal ini memastikan bahwa metode ini tidak melewatkan kombinasi yang mungkin menghasilkan performa terbaik.

**Penjelasan Setiap Hyperparameter**

Berikut adalah penjelasan dari setiap hyperparameter dalam `param_grid`:

1. **`n_estimators`**  
   Hyperparameter ini menentukan jumlah pohon keputusan (trees) yang akan dibangun oleh model LGBM Classifier (LightGBM). Semakin banyak pohon, semakin tinggi kapasitas model untuk menangkap pola dalam data. Namun, jumlah pohon yang terlalu sedikit dapat menyebabkan model underfitting karena tidak cukup kompleks untuk merepresentasikan data. Di sisi lain, jumlah pohon yang terlalu banyak dapat menyebabkan overfitting dan memperpanjang waktu pelatihan. Dalam grid ini, `n_estimators` diuji pada nilai [100, 200, 300] untuk mengeksplorasi keseimbangan antara kompleksitas model dan waktu pelatihan.

   Default: 100

2. **`learning_rate`**  
   Learning rate mengontrol seberapa besar langkah pembaruan model selama setiap iterasi pelatihan. Nilai yang kecil, seperti 0.01, membuat pelatihan lebih lambat tetapi cenderung lebih stabil, sedangkan nilai yang lebih besar, seperti 0.1, mempercepat pelatihan tetapi dapat melewatkan solusi optimal. Hyperparameter ini penting untuk memastikan bahwa model belajar secara efisien tanpa over- atau under-adjusting terhadap error.

   Default: 0.1

3. **`max_depth`**  
   Hyperparameter ini mengatur kedalaman maksimum pohon keputusan. Kedalaman pohon yang lebih kecil, seperti 3, membantu mencegah overfitting dengan membuat model lebih sederhana. Namun, kedalaman yang lebih besar, seperti 7, memungkinkan model menangkap pola yang lebih kompleks tetapi berisiko overfitting jika datanya terlalu sedikit atau noisy. Dalam grid ini, `max_depth` diuji pada nilai [3, 5, 7].

   Default: -1

4. **`subsample`**  
   Parameter ini menentukan proporsi sampel data yang digunakan untuk melatih setiap pohon. Nilai yang lebih kecil, seperti 0.8, membuat model lebih tahan terhadap overfitting dengan memaksa pohon belajar hanya dari sebagian data. Nilai 1.0 berarti semua data digunakan untuk melatih setiap pohon. Dengan menguji nilai [0.8, 1.0], kita dapat mengeksplorasi trade-off antara generalisasi dan kekuatan model.

   Default: 1.0

5. **`feature_fraction`**  
   Parameter ini menentukan proporsi fitur yang dipilih secara acak untuk membangun setiap pohon. Misalnya, nilai 0.8 berarti hanya 80% fitur yang digunakan untuk setiap pohon, yang dapat membantu mengurangi overfitting. Nilai 1.0 memungkinkan penggunaan semua fitur, yang bisa menangkap lebih banyak pola tetapi juga meningkatkan risiko overfitting.

   Default: 1.0

6. **`min_split_gain`**  
   Parameter ini mengatur pengurangan loss minimum yang diperlukan untuk membuat pembagian node baru di LightGBM. Nilai yang lebih tinggi, seperti 0.1 atau 0.2, menambahkan regularisasi pada pembagian node. Ini membantu memastikan bahwa hanya pembagian yang memberikan peningkatan signifikan pada loss yang dibuat, sehingga menghindari pembagian yang tidak penting atau overfitting.

   Default: 0

7. **`min_child_samples`**  
   Parameter ini menentukan jumlah minimum sampel yang diperlukan pada leaf node. Nilai yang lebih kecil, seperti 1, memungkinkan leaf node sangat kecil, sehingga model lebih kompleks. Sebaliknya, nilai yang lebih besar, seperti 5, mencegah leaf node kecil, membuat model lebih sederhana dan mencegah overfitting pada data yang noisy.

   Default: 20

8. **`scale_pos_weight`**
   
   Parameter scale_pos_weight pada LightGBM digunakan untuk menangani ketidakseimbangan kelas dalam dataset. Nilai ini menentukan bobot tambahan untuk kelas positif selama pelatihan. Nilai default scale_pos_weight pada LightGBM adalah 1.0, yang berarti tidak ada penyesuaian antara kelas positif dan negatif secara default. Untuk data yang tidak seimbang, di mana kelas positif hanya 9% dari total data, dapat dihitung dengan (scale_pos_weight = JumlahSampelNegatif / JumlahSampelPositif)
   
   Jika 91% data adalah negatif dan 9% positif maka scale_pos_weight = 91/9  10.11. Bisa ditetapkan scale_pos_weight = 10 untuk kasus ini.

   Default: 1.0

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold

# Define the transformer
transformer = ColumnTransformer([
    ('scaling', RobustScaler(), ['no_of_trainings', 'age', 'length_of_service', 'avg_training_score']),
    ('onehot', OneHotEncoder(drop='first'), ['education', 'gender', 'recruitment_channel']),
    ('binary', BinaryEncoder(), ['department', 'region'])
], remainder='passthrough')

# Define the parameter grid for LGBM Classifier (LightGBM)
param_grid = {
    'model__n_estimators': [90, 150],
    'model__learning_rate': [0.01, 0.2],
    'model__max_depth': [9, 12],
    'model__subsample': [0.6, 0.8, 1],
    'model__feature_fraction': [0.6, 0.8],
    'model__min_split_gain': [0, 0.5],
    'model__min_child_samples': [2, 4, 6],
    'model__scale_pos_weight':[10,20]
}

# Initialize LightGBM model
lgbm_model = lgb.LGBMClassifier()

# Create pipeline
pipeline = Pipeline([
    ('preprocess', transformer),
    ('model', lgbm_model)
])

# Cross-validation
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Define custom F-beta score with beta=0.5
f_beta_scorer = make_scorer(fbeta_score, beta=0.3)

# Grid Search for best parameters
grid_search = GridSearchCV(estimator=pipeline, param_grid=param_grid, scoring=f_beta_scorer, cv=skf, verbose=1, n_jobs=-1)
grid_search.fit(x_train, y_train)

# Best parameters and model
best_params = grid_search.best_params_
best_model_hyperparams = grid_search.best_estimator_

# Print results
print("Best Parameters:", best_params)
print("Best F-Beta Score (Cross-Validated):", grid_search.best_score_)
```

### Limitasi Model:

**Fitur Kategorikal**: Model akan terbatas pada kategori yang sudah ada dalam data pelatihan. Jika ada kategori baru yang tidak ada pada data pelatihan, seperti departemen atau region baru, model tidak akan tahu bagaimana menanganinya dan akan memberikan hasil yang tidak akurat.
